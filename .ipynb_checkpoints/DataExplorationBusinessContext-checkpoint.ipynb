## 2. Descripcin de los datos

Primero se lleva a cabo el recuento de datos y se revisa el tipo de datos que fueron importados utilizando el comando `data.info()`

```python
data.info(verbose=True)
```

```python
# Calidad de los datos
# Por facilidad de manipulacin y estandarizacin, los nombres de columnas que estan en PascalCase o separados por espacio 
# como el Hair color sern pasados a snake_case.

snake_case_names = {"Hair color": "hair_color",
                    "Pregnancies": "pregnancies",
                    "Glucose": "glucose",
                    "BloodPressure": "blood_pressure",
                    "City": "city",
                    "SkinThickness": "skin_thickness",
                    "Insulin": "insulin",
                    "DiabetesPedigreeFunction": "diabetes_pedigree_function",
                    "Age": "age",
                    "Outcome": "outcome"}
data = data.rename(columns=snake_case_names)


data.dtypes
```

Como se puede ver, tenemos solamente columnas de tipo string (corresponden a tipo object en pandas), e int (corresponde a int64 de numpy).

Sin embargo, las columnas de: Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin y DiabetesPedigreeFunction registran un tipo object aunque su contenido en realidad es int. Esto indica que tienen un placeholder string para cuando no se registr un valor. Debemos convertirlas a int, teniendo en cuenta la conversin de los placeholders a NaN.

Primero convertimos dichas columnas, que deberan ser numricas, a float64. Los datos que sean placeholders en string se convertiran en NaN. Luego nos desharemos de las columnas con datos que sean NaN para garantizar la calidad de los datos.

```python
# Conversin de columnas mencionadas anteriormente a float64, usando coercin de errores (datos no numricos)
headers = ["pregnancies", "glucose", "blood_pressure", "skin_thickness", "insulin", "diabetes_pedigree_function", "outcome"]
for header in headers:
    data[header] = pd.to_numeric(data[header], errors="coerce")
```

Despus de hacer el anlisis de cada una de las columnas, se nota que hay datos errneos para los siguientes criterios:

* Glucosa
* Presin sangunea
* Ancho de la piel
* Insulina

Para estos criterios se har una imputacin de los datos que aparecen como 0. Para ello, primero se sacar una copia de los datos y se reemplazarn aquellos que tienen valor 0 con NaN.

```python
data_copy = data.copy(deep = True)
data_copy[['glucose','blood_pressure','skin_thickness','insulin','BMI']] = data[['glucose','blood_pressure','skin_thickness','insulin','BMI']].replace(0,np.NaN)

## showing the count of Nans
print(data_copy.isnull().sum())
```

El siguiente paso es entender la distribucin de los datos para poder entender si es mejor imputarlos hacia la media o la mediana.

Aquellos datos cuya distribucin se asemeje a una normal pueden imputarse hacia la media, mientras que las dems distribuciones deben imputarse a la mediana.

```python
p = data.hist(figsize = (20,20))
```

```python
data_copy['glucose'].fillna(data_copy['glucose'].mean(), inplace = True)
data_copy['blood_pressure'].fillna(data_copy['blood_pressure'].mean(), inplace = True)
data_copy['skin_thickness'].fillna(data_copy['skin_thickness'].median(), inplace = True)
data_copy['insulin'].fillna(data_copy['insulin'].median(), inplace = True)
data_copy['BMI'].fillna(data_copy['BMI'].median(), inplace = True)
```

Las nuevas distribuciones de los datos se muestran a continuacin

```python
p = data_copy.hist(figsize = (20,20))
```

##### Cmo determinar el mejor modelo?

Para el caso actual, probaremos 3 modelos diferentes: MLP (Red neuronal), rbol de decisin, y KNN.

Ahora bien, el problema de negocio a tratar es la prediccin de diabetes en pacientes, dado un conjunto de parmetros. 

En este caso, la gravedad de las predicciones incorrectas no tiene el mismo impacto para el negocio.
Un falso positivo para este campo de estudio es un caso que el modelo predice como diabtico pero en realidad no lo es. 
Un falso negativo es un caso que el modelo predice como sano, pero en realidad es diabtico. 

Podemos afirmar que un falso positivo es un resultado menos perjudicial que un falso negativo, dado que si a un paciente enfermo de diabetes se le informa que est sano, corre grandes riesgos de salud. En cambio, un falso positivo podr descartarse sin poner en riesgo su salud, simplemente mediante la realizacin de otro examen clnico para reforzar el diagnstico.

Con ello en mente, las mtrica de desempeo que utilizaremos para escoger el mejor modelo, teniendo en cuenta el contexto de negocio, es el recall.
